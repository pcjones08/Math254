```r
# Recode the status variable into 1's and 0's.

mydata$cancer_status <- recode(mydata$status, 'M' = 1, 'B' = 0)

table(mydata$status)
table(mydata$cancer_status)
```

```r
mydata %>%
ggplot() +
  aes( x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5,)
```

```r
# prompt: can you add the logistic regression curve into my plot and color the curve as red

# Fit the logistic regression model
model <- glm(cancer_status ~ radius, data = mydata, family = "binomial")

# Create a sequence of radius values for prediction
radius_seq <- seq(min(mydata$radius), max(mydata$radius), length.out = 100)

# Predict probabilities for the radius sequence
predictions <- predict(model, newdata = data.frame(radius = radius_seq), type = "response")

# Add the logistic regression curve to the plot
mydata %>%
  ggplot() +
  aes(x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5) +
  geom_line(data = data.frame(radius = radius_seq, probability = predictions),
            aes(x = radius, y = probability), color = "red")

# Extract the coefficients from the model
coefficients <- coef(model)

# Create the equation string
equation <- paste0("y = ", round(coefficients[1], 3), " + ", round(coefficients[2],3), "*x")

# Add the equation to the plot
mydata %>%
ggplot() +
  aes(x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5) +
  geom_line(data = data.frame(radius = radius_seq, probability = predictions),
            aes(x = radius, y = probability), color = "red") +
  annotate("text", x = 10, y = 0.9, label = equation)
```

```r
mymodel <- glm(cancer_status ~ radius, data = mydata, family = binomial)
summary(mymodel)
```

Fitted curve:

$ \hat p  = \frac{e^{-15.2 + 1.03radius}}{1 + e^{-15.2 + 1.03radius
}}$

where $\hat p $ is the predicted probability from this model.

How to interpret the 'slope'?

- First, acknowledge the sign of the 'slope' parameter. In this case, the positive sign of the 'slope' indicates that probability increases as we move from small values to larger values (radii).

- In order to think about rate of change in probability, we need to make some judgements on the x values we use.


```r
mynewdata <- data.frame(radius = c(14,15, 20,21))

mypredictions <- predict(mymodel, newdata = mynewdata, type = 'response')

mypredictions
```
